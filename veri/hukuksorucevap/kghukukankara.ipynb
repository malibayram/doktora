```python
import requests
from bs4 import BeautifulSoup

import pandas as pd
```

```python
def temizle(metin):
  metin = metin.replace("\n", " ").replace("\t", " ").replace("\r", " ")
  while "  " in metin:
    metin = metin.replace("  ", " ")
  return metin.strip()
```

```python
response = requests.get("http://www.kghukukankara.com/sorucevap.aspx")
html = response.content
soup = BeautifulSoup(html, "html.parser")
soup
```

```python
div_col = soup.find("div", {"class": "col"})
list(div_col.children)
```

```python
""" 
'\n',
 <blockquote class="with-borders">
 <p>evim silahla basıldı 15 kişi tarafından benim olay yerinde olmamama rağmen benimde ismim verilmiş
 </p>
 <footer>m***** <cite title="Source Title">6/6/2024 9:42:19 AM</cite></footer>
 </blockquote>,
 '\n',
 <div class="testimonial testimonial-style-3">
 <blockquote>
 <p class="mb-0">Merhaba. 
 Konu ile ilgili yüz yüze görüşmemiz halinde yardımcı olabiliriz. Dosyanızı inceledikten sonra net cevap verebiliriz. </p>
 </blockquote>
 <div class="testimonial-arrow-down"></div>
 <div class="testimonial-author">
 <p><strong class="font-weight-extra-bold">KG Hukuk Danışmanlık</strong><span>Çankaya / Ankara</span></p>
 </div>
 </div>,
 '\n',
 <div class="divider divider-secondary">
 <i class="fas fa-chevron-down"></i>
 </div>,
 '\n',
 <blockquote class="with-borders"> """
```

```python
sorular = []

soru = {}
for child in div_col.children:
  # check if child is a blockquote and has cite tag inside
  if child.name == "blockquote" and child.find("cite"):
    # if yes, it means it is a question block
    soru['soru'] = temizle(child.find("p").text)
    soru['tarih'] = temizle(child.find("cite").text)
  # check if child is a div and has testimonial class
  elif child.name == "div" and "testimonial" in child.get("class", []):
    # if yes, it means it is an answer block
    soru['cevap'] = temizle(child.find("blockquote").text)
    sorular.append(soru)
    soru = {}
```
