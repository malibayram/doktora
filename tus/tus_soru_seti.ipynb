```python
import pandas as pd
import json
```

```python
# tipakademi.json
with open('tipakademi.json') as f:
    data = json.load(f)

len(data), data[100]
```

```python
# spread secenekler in each question
for i in range(len(data)):
    secenekler_dict = data[i]['secenekler']
    secenekler_list = list(secenekler_dict.keys())
    for j in range(len(secenekler_list)):
        data[i]["secenek" + secenekler_list[j]] = secenekler_dict[secenekler_list[j]]
    del data[i]['secenekler']
len(data), data[100]
```

```python
tus_df = pd.DataFrame(data)
# tus_df.to_csv('tus.csv', index=False, encoding='utf-8')
tus_df
```

```python
# add new column as source and fill with 'tipakademi'
tus_df['source'] = 'tipakademi.json'
tus_df.to_csv('tus.csv', index=False, encoding='utf-8')
tus_df
```

```python
# read all json files inside tus_resimli/sorular
import os

tus_resimli_sorular = []
for filename in os.listdir('tus_resimli/sorular'):
    if filename.endswith('.json'):
      with open('tus_resimli/sorular/' + filename) as f:
          data = json.load(f)
          for soru_dict in data:
              soru = {}
              soru['soruSirasi'] = soru_dict['sira']
              soru['dogruCevap'] = soru_dict['yanit']
              if 'soruUstKismi' in soru_dict:
                soru['soruUstKismi'] = soru_dict['soruUstKismi']
              try:
                 soru['soru'] = soru_dict['soru']
              except:
                 try:
                    soru['soru'] = soru_dict['soru ']
                 except:
                    print(soru_dict)
              soru['secenekA'] = soru_dict['a']
              soru['secenekB'] = soru_dict['b']
              soru['secenekC'] = soru_dict['c']
              soru['secenekD'] = soru_dict['d']
              soru['secenekE'] = soru_dict['e']
              if 'aciklama' in soru_dict:
                soru['aciklama'] = soru_dict['aciklama']
              soru['source'] = filename
              tus_resimli_sorular.append(soru)
len(tus_resimli_sorular), tus_resimli_sorular[0]
```
