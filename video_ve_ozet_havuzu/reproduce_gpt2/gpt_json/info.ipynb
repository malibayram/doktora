```python
with open("info.md", "r") as f:
    lines = f.readlines()


# each line - filename: sd_hf_transformer.h.9.attn.c_attn.weight.json torch.Size([768, 2304])
# then add to the and of the list as | sd_hf_transformer.h.0.mlp.c_proj.bias | torch.Size([768]) |

for i in range(len(lines)):
    line = lines[i]
    if line.startswith("- filename: "):
        filename = line.split()[2]
        # after filename there is size as torch.Size([768, 2304])
        # size should be torch.Size([768, 2304])
        size = line.split()[3:]
        size = " ".join(size)
        
        lines.append(f"| {filename} | {size} |")

for line in lines:
    print(line)

# write to file
with open("info2.md", "w") as f:
    for line in lines:
        f.write(line+"\n")
```
