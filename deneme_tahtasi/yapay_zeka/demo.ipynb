```python
w1 = 0.70
w3 = 0.40
def z1(x, y):
  return x * w1 + y * w3

noktalar = set()
for i in range(10):
  for j in range(10):
    x = i
    y = j
    z = z1(x, y)
    noktalar.add((x, y, z))

print(len(noktalar))
      
```

```python
# draw the points
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
for x, y, z in noktalar:
  ax.scatter(x, y, z, c='r', marker='o')
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')
plt.show()
```

```python
import plotly.graph_objects as go
import numpy as np

# Parameters and function
w1 = 0.7
w2 = 0.3
w3 = 0.4
w4 = 0.6

def z1(x, y):
    return x * w1 + y * w3

def z2(x, y):
    return x * w2 + y * w4

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Generate points
noktalar = set()
for i in range(20):
    for j in range(20):
        x = i - 10
        y = j - 10
        z = z1(x, y)
        z = sigmoid(z)
        noktalar.add((x, y, z))

# Unpack points
x_vals, y_vals, z_vals = zip(*noktalar)

# Create a 3D scatter plot
fig = go.Figure(data=[go.Scatter3d(
    x=x_vals,
    y=y_vals,
    z=z_vals,
    mode='markers',
    marker=dict(size=5, color=z_vals, colorscale='Viridis', opacity=0.8)
)])

# Add layout details
fig.update_layout(
    scene=dict(
        xaxis_title='X Axis',
        yaxis_title='Y Axis',
        zaxis_title='Z Axis'
    ),
    title="Dynamic 3D Points Visualization",
    width=800,
    height=800
)

# Show the plot
fig.show()
```

```python
import plotly.graph_objects as go

# Parameters and function
w1 = 0.70
w2 = 0.30
w3 = 0.40
w4 = 0.60

def z1(x, y):
    return x * w1 + y * w3

def z2(x, y):
    return x * w2 + y * w4

# Generate points
noktalar = set()
for i in range(10):
    for j in range(10):
        x = i
        y = j
        noktalar.add((x, y, z1(x, y)))
        noktalar.add((x, y, z2(x, y)))

# Unpack points
x_vals, y_vals, z_vals = zip(*noktalar)

# Create a 3D scatter plot
fig = go.Figure(data=[go.Scatter3d(
    x=x_vals,
    y=y_vals,
    z=z_vals,
    mode='markers',
    marker=dict(size=5, color=z_vals, colorscale='Viridis', opacity=0.8)
)])

# Add layout details
fig.update_layout(
    scene=dict(
        xaxis_title='X Axis',
        yaxis_title='Y Axis',
        zaxis_title='Z Axis'
    ),
    title="Dynamic 3D Points Visualization",
    width=800,
    height=800
)

# Show the plot
fig.show()
```
