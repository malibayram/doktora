```python
import pandas as pd
```

```python
data = pd.read_csv('generated_answers_llama3_8b_with_similarities_and_scores.csv')
data
```

```python
# change gpt3.5_score_generated_answer and gpt3.5_score_original_answer from 8/10 to 0.8

for i, row in data.iterrows():
    value1 = str(row['gpt3.5_score_generated_answer'])
    # if value1 contains a slash, take the first part
    if '/' in value1:
        value1 = value1.split('/')[0]
    value2 = str(row['gpt3.5_score_original_answer'])
    if '/' in value2:
        value2 = value2.split('/')[0]
    value1 = float(value1)
    value2 = float(value2)
    print(value1, value2)
    if value1 > 0:
        data.at[i, 'gpt3.5_score_generated_answer'] = value1 / 10
    if value2 > 0:
        data.at[i, 'gpt3.5_score_original_answer'] = value2 / 10
```

```python
data
```

```python
# get average of each column original_answer_to_generated_answer_similarity, original_answer_to_llama3_8b_answer_similarity, gpt3.5_score_generated_answer, gpt3.5_score_original_answer
oa_to_ga = data['original_answer_to_generated_answer_similarity'].mean()
oa_to_la = data['original_answer_to_llama3_8b_answer_similarity'].mean()
gpt3_5_ga = data['gpt3.5_score_generated_answer'].mean()
gpt3_5_oa = data['gpt3.5_score_original_answer'].mean()
```

```python
print("Orjinal cevabın bizim modelin ürettiği cevaba benzerlik ortalaması:", oa_to_ga)
print("Orjinal cevabın llama3_8b modelin ürettiği cevaba benzerlik ortalaması:", oa_to_la)
print("Orjinal cevaba GPT-3.5'in verdiği puan ortalaması:", gpt3_5_ga)
print("Bizim modelin ürettiği cevaba GPT-3.5'in verdiği puan ortalaması:", gpt3_5_oa)

benzerlik_skoruna_gore_gelisme = round((oa_to_ga - oa_to_la) / oa_to_la * 100, 2)
gpt3_5_skoruna_gore_fark = round((gpt3_5_ga - gpt3_5_oa) / gpt3_5_oa * 100, 2)
print("Orjinal cevabın bizim modelin ürettiği cevaba benzerlik ortalamasındaki gelişme oranı:", benzerlik_skoruna_gore_gelisme, "%")
print("Bizim modelin ürettiği cevaba GPT-3.5'in verdiği puan ortalamasındaki fark oranı:", gpt3_5_skoruna_gore_fark, "%")
```

```python
scores_dict = {
    "Orjinal cevabın bizim modelin ürettiği cevaba benzerlik ortalaması": oa_to_ga,
    "Orjinal cevabın llama3_8b modelin ürettiği cevaba benzerlik ortalaması": oa_to_la,
    "Orjinal cevaba GPT-3.5'in verdiği puan ortalaması": gpt3_5_ga,
    "Bizim modelin ürettiği cevaba GPT-3.5'in verdiği puan ortalaması": gpt3_5_oa,
    "Orjinal cevabın bizim modelin ürettiği cevaba benzerlik ortalamasındaki gelişme oranı": benzerlik_skoruna_gore_gelisme,
    "Bizim modelin ürettiği cevaba GPT-3.5'in verdiği puan ortalamasındaki fark oranı": gpt3_5_skoruna_gore_fark
}

scores_dict_df = pd.DataFrame(scores_dict.items(), columns=['Metric', 'Value'])
scores_dict_df.to_csv('scores_dict.csv', index=False)
scores_dict_df
```
