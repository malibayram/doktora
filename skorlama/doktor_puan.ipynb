```python
import json
import pandas as pd
```

```python
doktor_puanlar_json = open("doktor_puanlar.json")
doktor_puanlar = json.load(doktor_puanlar_json)

doktor_puanlar_json.close()

doktor_puanlar_df = pd.DataFrame(doktor_puanlar)
doktor_puanlar_df
```

```python
# remove rows if cevaplayan column is not Nan
# doktor_puanlar_df = doktor_puanlar_df[doktor_puanlar_df["cevaplayan"].isnull()]
doktor_puanlar_df = doktor_puanlar_df.drop(columns=["cevaplayan", "referans"])
doktor_puanlar_df
```

```python
# check if there are any duplicate rows by checking if there are any duplicate values in the columns of puanlayanId and tarih
doktor_puanlar_df[doktor_puanlar_df.duplicated(subset=["puanlayanId", "tarih"])]
doktor_puanlar_df = doktor_puanlar_df.drop_duplicates(subset=["puanlayanId", "tarih"])
doktor_puanlar_df
```

```python
# count cevapModel column value if it is equal to doktorllama3-8b

doktor_puanlar_df[doktor_puanlar_df["cevapModel"] == "doktorllama3-8b"].count(), doktor_puanlar_df[doktor_puanlar_df["cevapModel"] == "doktor"].count(), doktor_puanlar_df[doktor_puanlar_df["cevapModel"] == "llama3-8b"].count()
```

```python
# get puan column mean of each cevapModel value

group = doktor_puanlar_df.groupby("cevapModel")
group["puan"].mean()
```

```python
doktor_puanlar_df.to_csv("doktor_puanlar.csv", index=False)
```
