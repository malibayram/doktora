```python
from sentence_transformers import SentenceTransformer
```

```python
# Load https://huggingface.co/sentence-transformers/all-mpnet-base-v2
model = SentenceTransformer("paraphrase-multilingual-mpnet-base-v2")
```

```python
embeddings1 = model.encode([
    "Adetin 3. günü hormon tahlili ve ultrason ile yumurtalıkların değerlendirilmesi gereklidir. Bunlara göre ya hemen tedaviye başlanır yada 6 ay daha beklenebilir. Geçmiş olsun.",
    "Bu konudaki en sağlıklı cevabı, nöroloji uzmanınız verebilir.",
    "Cok tesekkür ederim, sevgili kardeşim!",
    "Bu konudaki en sağlıklı cevabı, nefrologist verebilir."
])
embeddings2 = model.encode([
    "Merhaba, öncelikle muayene olmadan net bir şey söylemek mümkün değil ama adetleriniz düzenli olduğu halde gebe kalamıyor iseniz eşinizin spermiyogram yaptırması ve sizin de histereskopi yaptırmanız gerekebilir. Ama sizden hormon tetkiki istemişlerdir muhtemelen, o da normalse eğer ultrason ve muayene ile rahim içi görüntülenmesinizi öneririm. Çok geçmiş olsun.",
    "Merhaba, öncelikle geçmiş olsun. Bu tür yaralanmalarda nörolojik gelişme önemlidir ve bu hastalar 6 ay ile 1 yıl içerisinde düzelme gösterirler. Bu süreçte fizik tedavi de önemli rol oynar. Benim önerim kardeşinize en yakın Fizik Tedavi ve Rehabilitasyon Merkezine gitmenizdir. Sağlıklı günler dileriz.",
    "Ancak, kardeşimin bu durumunu daha fazla detayandırarak tedavi planını oluşturmamız gerekir.",
    "Bu konudaki en sağlıklı cevabı, nöroloji uzmanınız verebilir."
])
similarities = model.similarity(embeddings1, embeddings2)
similarities
```
