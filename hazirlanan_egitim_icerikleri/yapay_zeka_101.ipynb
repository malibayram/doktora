```python
# draw atis point on the plot

def atisi_ciz(atis):
    fig, ax = plt.subplots()
    circle = plt.Circle((0, 0), 30, color='b', fill=False)
    ax.add_artist(circle)
    ax.set_xlim(-100, 100)
    ax.set_ylim(-100, 100)
    plt.gca().set_aspect('equal', adjustable='box')
    plt.scatter(0,0, color='b')
    plt.scatter(*atis, color='r')
    plt.show()
```

```python
atisi_ciz(atis)
```

```python
yari_cap = 30
```

```python
def tahmin_et(atis):
  x = atis[0]
  y = atis[1]

  mesafe = np.sqrt(x**2 + y**2)

  return mesafe < yari_cap
  
```

```python
atislar = []
for i in range(200):
  atis = (np.random.randint(-100, 100), np.random.randint(-100, 100))
  # atisi_ciz(atis)
  tahmin = tahmin_et(atis)
  atislar.append({
    "x": atis[0],
    "y": atis[1],
    "isabet": tahmin
  })
  print(atis, tahmin)
```

```python
np.random.seed(0)
random_yari_cap = np.random.randint(0, 100)
type(random_yari_cap)
```

```python
ogrenme_orani = 0.01
```

```python
def yeni_tahmin_et(atis):
  x = atis[0]
  y = atis[1]

  mesafe = np.sqrt(x**2 + y**2)

  return mesafe < random_yari_cap
```

```python
for epoch in range(1, 200):
  for i in range(atislar_df.shape[0]):
    atis = atislar_df.iloc[i]
    tahmin = yeni_tahmin_et((atis["x"], atis["y"]))
    hata = int(atis["isabet"]) - int(tahmin)
    random_yari_cap = random_yari_cap + hata * ogrenme_orani  
    # print(atis["x"], atis["y"], tahmin, atis["isabet"])

  print(f"{epoch}. epoch sonrasƒ± yarƒ± √ßap: {random_yari_cap}")
```
